>* `Decision Trees` can perform both classification `DecisionTreeClassifier` and regression `DecisionTreeRegressor` tasks, and even multioutput task.
>* Here is the code to build it:

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris(as_frame = True) # to return the data as dataframe not array or sparce-matrix
X_iris = iris.data[['petal length (cm)','petal width (cm)']].values
y_iris = iris.target

tree_clf = DecisionTreeClassifier(max_depth =2 , random_state =22)
tree_clf.fit(X_iris, y_iris)
```

>* You can visualize this tree by the following:
>>* Export the `tree_clf` as .dot file "file written in DOT language for visualizing graphs" using the `export_graphviz` function.
>>* Use `graphviz` library to load the `.dot` file and dispay it into the jupyter notebook as graph.


```python
from sklearn.tree import export_graphviz

export_graphviz(
    tree_clf,
    out_file = 'iris_tree.dot',
    feature_names = ["petal length (cm)", "petal width (cm)"],
    class_names = iris.target_names,
    rounded = True,
    filled = True
)

```

```python
from graphviz import Source

Source.from_file("iris_tree.dot")
```

#### Analyzing the above tree:

>* First, you start at the **root node** and ask if your `petal length` is greater than 2.45 "This is (depth 0)"
>* If it is True, then go to the left child node technically it is a **leaf-node** (means there is no children there).
>* If it is False, then go to the right child node ... because there is a condition there, this node is called **split-node**.
>>* If it is True, then go to the left ... and if it is False, then go to the right.
>
>

>* Decision Tree doesn't need much data preparation (like feature scaling in case of SVM).

>>* A node’s `samples` attribute counts how many training instances it applies to
>>* A node’s `value` attribute tells you how many training instances of each class this node applies to: for example, the bottom-right node applies to 0 Iris setosa, 1 Iris versicolor, and 45 Iris virginica.
>>* A node’s `gini` attribute measures its Gini impurity: a node is “pure” (gini=0) if all training instances it applies to belong to the same class. For example, since the depth-1 left node applies only to Iris setosa training instances, it is pure and its Gini impurity is 0.
>>* `gini` is calculated by this way:
The depth-2 left node has a Gini impurity equal to `1 – (0/54)2 – (49/54)2 – (5/54)2 ≈ 0.168`


>* Scikit-Learn uses the `CART` algorithm, which produces only binary trees, meaning trees where split nodes always have exactly two children (i.e., questions only have yes/no answers)
>* However, other algorithms, such as `ID3`, can produce decision trees with nodes that have more than two children.

For the above tree, the decision boundaries may be like the following images (plotting the two features against each other):
>* The **SolidLine** is for the split-node at **depth 0**.
>* The **dashedLine** is for the split-node at **depth 1**.
>* The **dottedLine** is actually not for this tree but it assumes if we have split-node at the **depth 2** it will be like this.

Training the tree:

>* `SckitLearn` uses the CART algorithm to train the tree ... this algorithm tries to get the pair (`k` "one feature",`tk` "one threshold") that achieves the split with the minimum the least impurities weighted by their size.
>* The cost function for specific split node is: 

>* Note this impurity has two measures (`Gini` "**default measure**", `Entropy` "another measure").

>* it splits each split nodes using the same logic, then the sub-subsets, and so on, recursively.
>* It stops if:
>>* It reaches the max depth. or
>>* There is no split that reduces the impurity.
>>* There are other hyperparameters that control the stopping condition: `min_samples_split`, `min_samples_leaf`, `min_weight_fraction_leaf`, and `max_leaf_nodes`.
>* Note the CART algorithm is a greedy one, means it produces a solution that’s reasonably good **but not guaranteed to be optimal**. (decision tree tries to find the optimal split at the current split but doesn't but into consideration the following splits together produce the general optimal split or not.)
_____________________________
>>* For the `Gini`, it is calculated like this:
>>>* It is calculated by this way: The depth-2 left node has a **`Gini`** impurity equal to `1 – (0/54)2 – (49/54)2 – (5/54)2 ≈ 0.168`
>>>
>>>> The Matrix Form of this cost function (`I` is the identity matrix):
 
>>* For the `entropy`, there are multiple definitoins:
>>>* Entropy approaches zero when molecules are still and well ordered.
>>>* If you measure the average information content of a message, Entropy is zero when all messages are identical.
>>>* As an impurity measure, entropy is zero when it contains instances of only one class.
>>>* For example, the depth-2 left node in the above tree has an **`entropy`** equal to `–(49/54) log2 (49/54) – (5/54) log2 (5/54) ≈ 0.445`
>>>* <font color = red>Note: `log2(0)` is undefined ... Therefore, put `log2(0.000001)` for correct computations</font> if you build the algorithm from scratch.
>>> 
>>>* Andrew Note about `Entropy`. The entropy of p = 1/3 (one dog and two cats out of 3 animals) is the same as p = 2/3 (one dog and two cats out of 3 animals):
>>> 
>>>* There is a more edited formula of the Entropy called the `information gain`. "We try to minimize the entropy but maximize the information gain":
>>>
>>>* P_root measures the proportion of the each class in our dataset.(here we consider the propotion of the cat). 
>>>
>>>* We do all the above in order to decide which feature should we split on. We have many trees each with different feature, **we try to choose the feature that achieves the maximum information gain.**
>>>* To put it all together:
>>> 
>>* The two measure has no large difference but:
>>>* `Gini` is faster to compute. But `Entropy` sometimes is better in case of imbalanced data.(مش متأكد أوى)

>* Unfortunately, **finding the optimal tree** is known to be an `NP-complet`e problem. It requires **`O(exp(m))`** time, making the problem intractable even for small training sets. This is why we must settle for a “reasonably good” solution when training decision trees.
______________________

>* Once you built the decision tree and set each node information, the **`prediction`** complexity of the prediction you want to do is `O(log2(m))` where m is the number of nodes.

>* But in **`training`** the tree and trying to find the nodes information, the complexity is `O(n x m log2(m))`. where n is the number of features and m is the number of nodes.
________________________

>* DecisionTree, if it is left without restrictions, will overfit the training examples because it will adapt itself to the classes.
>* The above is a drawback of non-parametric models.

Parameteric Vs Non-parameteric models:

>* The `Parametric` model is the model **you <font color = 'red'> know </font>the number of its parameters before training** like LinearRegression, SVM, NB..etc. (by knowing the number of parameters, you have limited degree of freedom which decrease the risk of overfitting but increase the risk of underfitting).
>* The `Non-Parametric` model is the model **you <font color = 'red'>don't know</font> the number of its parameters before training** like DecisionTree. (This hs unlimited degree of freedom which increases the risk of overfitting).

>* To avoid overfitting the training data, you need to restrict the decision tree’s freedom during training. The regularization hyperparameter depends on the training algorithm used (either CART, ID3 ...etc).
>* But in general, you can regularize the tree and reduce its degree of freedom by reducing the `max_depth`.

>* These are more parameters for regularizing other than `max_depth`:
>>
>*
>> 
_______________


```python
# Regularization in action:

from sklearn.datasets import make_moons

X_moons, y_moons = make_moons(n_samples = 150, noise=0.2, random_state=42)
tree_clf1 = DecisionTreeClassifier(random_state=22) # without regularization at all
tree_clf2 = DecisionTreeClassifier(max_depth = 3, random_state=22)
tree_clf3 = DecisionTreeClassifier(min_samples_leaf = 5, random_state=42)

tree_clf1.fit(X_moons,y_moons)
tree_clf2.fit(X_moons,y_moons)
tree_clf3.fit(X_moons,y_moons)
```

```python
X_test,y_test = make_moons(n_samples=1000,noise=0.2,random_state =21)

print("first: ", tree_clf1.score(X_test,y_test) )
print("second: ",tree_clf2.score(X_test,y_test) )
print("third: ",tree_clf3.score(X_test,y_test)  )
```

```python
export_graphviz(tree_clf1,
                out_file='tree_clf1.dot')
export_graphviz(tree_clf2,
                out_file='tree_clf2.dot')
export_graphviz(tree_clf3,
                out_file='tree_clf3.dot')
```

### Regression in decision trees:
 
>* Decision trees can be used for regression by doing the same splits technique using trainning algorithms like CART (**but the splitting crieteria is the `mse` not the `gini`**)
>

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor 
np.random.seed(22)

np.random.seed(42)
X_quad = np.random.rand(200, 1) - 0.5  # a single random input feature
y_quad = X_quad ** 2 + 0.025 * np.random.randn(200, 1)

tree_reg = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg.fit(X_quad, y_quad)
```

### Drawbacks of Decision Trees:

>* 1. **Decision trees love orthogonal decision boundaries** (all splits are perpendicular to an axis), which makes them sensitive to the data’s orientation.
>>* One way to solve this is to `scale the data` then `apply PCA`. (the resulted component is used as feature to plot the data against it and split it).
>>>* We will look at PCA in detail in Chapter 8, but for now you only need to know that it rotates the data in a way that reduces the correlation between the features, which often (not always) makes things easier for trees. As you can see, the rotation makes it possible to fit the dataset pretty well using only one feature, z1, which is a linear function of the original petal length and width
>* 

>* 2. The main issue with decision trees is that they have quite a `high variance`: small changes to the hyperparameters or to the data may produce very different models. In fact, since the training algorithm used by Scikit-Learn is stochastic—it randomly selects the set of features to evaluate at each node—even retraining the same decision tree on the exact same data may produce a very different model (unless you set the random_state hyperparameter).
>>* To solve this problem: by averaging predictions over many trees, it’s possible to reduce variance significantly. Such an **ensemble of trees** is called a `random forest`.

>* 3. **Decision is prone to overfitting**
>>* To solve this, we do regularization.


```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
# from category_encoders import OneHotEncoder

pca_pipeline = make_pipeline(StandardScaler(), PCA())
X_iris_rotated = pca_pipeline.fit_transform(X_iris)
tree_clf_pca = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clf_pca.fit(X_iris_rotated, y_iris)
```
