>* The `%matplotlib inline` command tells the IPython environment to draw the plots immediately after the current cell. The drawn plots are shown below the code and stored in the notebook document for future reference.
>* The line magic command `%matplotlib inline` enables the drawing of matplotlib figures in the IPython environment. Once this command is executed in any cell, then for the rest of the session, the matplotlib plots will appear directly below the cell in which the plot function was called.
>* The matplotlib inline command is greatly helpful as it distinguishes between plots of different cells. For each cell, the graph is plotted after the current cell which does not affect the graphs of previous cells.
>* It creates a separation between graphs between cells such that settings like color pallets if applied in the current cell, will not affect the plots of any previous cells.

>* In the **latest versions** of IPython and Jupyter notebooks, the command `%matplotlib inline` is *unnecessary as the plots are displayed below the cell by default*.
>* In **older versions** of IPython and Jupyter notebooks, *if the interactive mode is disabled and the graphs are plotted using matplotlib, they will not appear unless the `matplotlib.pyplot.show()` function is called.*

>* PyCharm is a **Python IDE**. The magic commands are available in the **IPython environment**; hence, *the matplotlib inline command is not available in PyCharm*.

>* There are also examples like : **`%matplotlib notebook`** (backend for interactive plot)

```python
# Creates blank canvas
fig = plt.figure()

# fig.add_axes( [left, bottom, width, height] )
axes1 = fig.add_axes([0.0, 0.0, 0.6, 0.6]) # main axes
axes2 = fig.add_axes([0.2, 0.2, 0.4, 0.4]) # inset axes
x= np.arange(0,7)
y= 2*x

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(x, x**2, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) 
ax2 = fig.add_axes([0.2,0.5,.2,.2])


def ax_plt(ax, x , y, xlim:list, ylim:list, style:str,
           x_label:str, y_label:str, title:str):
    
    ax.plot(x,y,style)
    
    ax.set_xlim(xlim[0],xlim[1])
    ax.set_ylim(ylim[0],ylim[1])
    
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title)
    
x = np.arange(0,100)
y = 2*x

ax_plt(ax1, x, y, [0,100], [0,200], 'r-', 'x', 'y', 'plot1' )
ax_plt(ax2, x, y, [0,100], [0,200], 'r-', 'x', 'y', 'plot2')
```

```python
plt.subplots(1,2,figsize=(11,2))

ax1 = plt.subplot(121)
x = np.arange(0,100)
y = 2*x
ax_plt(ax1, x, y, [0,100], [0,200], 'b-', 'X', 'Y', '1st')

ax2 = plt.subplot(122)
y = x**2
ax_plt(ax2, x, y, [0,100], [0,10_000], 'r--', 'X', 'Y', '2nd')
```

```python
# % of the sold for each type in the last month only 
trans_df["% last month"] =   round( (trans_df[11]  /   trans_df.sum(axis=1)) * 100 , 2)
trans_df["% last month"]
```

> **`autopct`** enables you to display the percent value using Python string formatting. For example, if autopct='%. 2f' , then for each pie wedge, the format string is '%. 2f' and the numerical percent value for that wedge is pct , so the wedge label is set to the string '%

```python
my_labels = df.iloc[:, 1:-2].columns
my_x = trans_df["% last month"][1:-2]

# without percentage
plt.figure(figsize=(16,8))
plt.subplot(121)
plt.pie(my_x ,labels = my_labels, autopct= '%.2f',
        shadow = True, explode = [0,0,0.2,0,0,0]);


# the percentage appears
plt.subplot(122)
plt.pie(my_x ,labels = my_labels, autopct= '%.2f%%',
       shadow = False, explode = [0.1,0,0,0,0,0]);
```
